Preprocess Bastidas-Ponce 2019 mouse pancreatic endocrinogenesis for CPM.

Follows CPM paper preprocessing for scRNA-seq lineage data (like mouse gastrulation, S4.3):
  1. Normalize to 10,000 counts per cell
  2. Log1p transform
  3. Top 2,000 highly variable genes

This dataset is not in the CPM paper — it's a novel test of the method.

Biological structure:
  Ductal (progenitor) -> Ngn3 low EP -> Ngn3 high EP -> Pre-endocrine ->
                                                         {Alpha, Beta, Delta, Epsilon}

```python
import scanpy as sc
import scvelo as scv
```

```python
OUTPUT_CSV = 'pancreas/pancreas.csv'

adata = scv.datasets.pancreas()
print(f'Raw: {adata.shape}')
print(f'Cell types: {adata.obs["clusters"].unique().tolist()}')
```

```python
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
sc.pp.highly_variable_genes(adata, n_top_genes=2000, flavor='seurat')
adata = adata[:, adata.var.highly_variable].copy()
print(f'After HVG filter: {adata.shape}')
```

```python
df = adata.to_df()
df['celltype'] = adata.obs['clusters'].values
df.to_csv(OUTPUT_CSV, index=False)
print(f'Wrote {OUTPUT_CSV}: {df.shape}')
print(f'Last col unique values: {df["celltype"].unique().tolist()}')
```
